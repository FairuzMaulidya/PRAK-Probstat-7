```python
import pandas as pd

# Membaca file CSV dengan delimiter titik koma
data_fairuz = pd.read_csv('C:/Users/Fairuz Maulidya/Documents/probstat/titanic.csv', delimiter=';')

# Menampilkan data
print(data_fairuz)
```

```python
#mengambil data pada kolom tertentu
data1 = data_fairuz.loc[:,['Age','Pclass','Survived']] 
print(data1)
```

```python
#memvisualisasikan data titanic
data2 = data_fairuz[['Age', 'Pclass', 'Survived']]
data2.plot(title='Persebaran Data', x='Age', y='Pclass', kind='scatter', c='Survived', colormap='Paired')
```

```python
#memanipulasi data jumlah penumpang berdasarkan group Pclass 
data3 = data_fairuz[['Name', 'Sex', 'Age', 'Pclass', 'Fare']] 
penumpang=data3.groupby('Pclass')['Name'].nunique() 
print('Jumlah Penumpang:\n', penumpang)
```

```python
#memfilter data penumpang yang selamat berdasarkan pclass 
data4 = data_fairuz[['Name', 'Sex', 'Age', 'Pclass', 'Fare']]
notsurvivedpassanger=data4['Pclass'].loc[data_fairuz['Survived']==0] 
print('Penumpang yang tidak survived:\n', notsurvivedpassanger.value_counts()) 
survivedpassanger=data4['Pclass'].loc[data_fairuz['Survived']==1] 
print('\nPenumpang yang survived:\n', survivedpassanger.value_counts())
```

```python
# Mengelompokkan data berdasarkan kolom "Sex" dan menghitung jumlah penumpang dalam setiap kelompok
jumlah_penumpang_per_sex = data_fairuz.groupby('Sex').size()

# Menampilkan hasil manipulasi data
print(jumlah_penumpang_per_sex)
```

```python
# Filter data penumpang yang selamat berdasarkan jenis kelamin
penumpang_selamat_per_sex = data_fairuz[(data_fairuz['Survived'] == 1) & (data_fairuz['Sex'] == 'female')]

# Menampilkan hasil filter data
print(penumpang_selamat_per_sex)
```
